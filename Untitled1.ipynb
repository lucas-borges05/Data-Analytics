```python

# Converter a coluna 'data' para datetime
dados_bitcoin['Data'] = pd.to_datetime(dados_bitcoin['Data'])
dados_ethereum['Data'] = pd.to_datetime(dados_ethereum['Data'])
dados_xrp['Data'] = pd.to_datetime(dados_xrp['Data'])
dados_cardano['Data'] = pd.to_datetime(dados_cardano['Data'])
dados_solana['Data'] = pd.to_datetime(dados_solana['Data'])

dados_bitcoin['Data'] = pd.to_datetime(dados_bitcoin['Data'])
dados_ethereum['Data'] = pd.to_datetime(dados_ethereum['Data'])
dados_xrp['Data'] = pd.to_datetime(dados_xrp['Data'])
dados_cardano['Data'] = pd.to_datetime(dados_cardano['Data'])
dados_solana['Data'] = pd.to_datetime(dados_solana['Data'])

# Plotando o volume de cada criptomoeda
plt.figure(figsize=(14, 7))

# Plotando cada criptomoeda
plt.plot(dados_bitcoin['Data'], dados_bitcoin['Volume'], label='Bitcoin', linewidth=2)
plt.plot(dados_ethereum['Data'], dados_ethereum['Volume'], label='Ethereum', linewidth=2)
plt.plot(dados_xrp['Data'], dados_xrp['Volume'], label='XRP', linewidth=2)
plt.plot(dados_cardano['Data'], dados_cardano['Volume'], label='Cardano', linewidth=2)
plt.plot(dados_solana['Data'], dados_solana['Volume'], label='Solana', linewidth=2)

# Ajuste no formato das datas para exibir apenas o ano
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Ajuste o intervalo para as datas para não sobrecarregar o gráfico
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Cada ano
plt.gca().xaxis.set_minor_locator(mdates.MonthLocator())  # Para ticks menores (não exibidos)

# Configurações do gráfico
plt.title('Volume de Criptomoedas ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)

# Ajustar a rotação das labels no eixo X para melhorar a legibilidade
plt.xticks(rotation=45, ha='right')

# Ajuste do layout para não cortar as labels
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
dados_bitcoin['Data'] = pd.to_datetime(dados_bitcoin['Data'])
dados_ethereum['Data'] = pd.to_datetime(dados_ethereum['Data'])
dados_xrp['Data'] = pd.to_datetime(dados_xrp['Data'])
dados_cardano['Data'] = pd.to_datetime(dados_cardano['Data'])


# Realizar o merge entre os DataFrames, com sufixos distintos para as colunas de volume
dados_merged = pd.merge(dados_bitcoin[['Data', 'Volume']], dados_ethereum[['Data', 'Volume']], on='Data', suffixes=('_bitcoin', '_ethereum'))
dados_merged = pd.merge(dados_merged, dados_xrp[['Data', 'Volume']], on='Data', suffixes=('_ethereum', '_xrp'))
dados_merged = pd.merge(dados_merged, dados_cardano[['Data', 'Volume']], on='Data', suffixes=('_xrp', '_cardano'))
dados_merged = pd.merge(dados_merged, dados_solana[['Data']], on='Data', suffixes=('_cardano',))

# Verificar as colunas do DataFrame após os merges
print(dados_merged.columns)
dados_merged.head()

```

```python
# Selecionar apenas as colunas de volume das criptomoedas
import seaborn as sns
dados_corr = dados_merged[['Volume_bitcoin', 'Volume_ethereum', 'Volume_xrp', 'Volume_cardano']]

# Calcular a matriz de correlação
correlation_matrix = dados_corr.corr()

# Plotar a matriz de correlação usando Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlação entre os Volumes das Criptomoedas')
```

```python
#mudar eixo x para ano, extraindo o Ano da Data

dados_bitcoin['Ano'] = dados_bitcoin['Data'].dt.year
volume_anual_bitcoin = dados_bitcoin.groupby('Ano')['Volume'].sum()

dados_ethereum['Ano'] = dados_ethereum['Data'].dt.year
volume_anual_ethereum = dados_ethereum.groupby('Ano')['Volume'].sum()

dados_xrp['Ano'] = dados_xrp['Data'].dt.year
volume_anual_xrp = dados_xrp.groupby('Ano')['Volume'].sum()

dados_cardano['Ano'] = dados_cardano['Data'].dt.year
volume_anual_cardano = dados_cardano.groupby('Ano')['Volume'].sum()

dados_solana['Ano'] = dados_solana['Data'].dt.year
volume_anual_solana = dados_solana.groupby('Ano')['Volume'].sum()
```

```python
volume_anual_bitcoin = dados_bitcoin.groupby('Ano')['Volume'].sum()
volume_anual_ethereum = dados_ethereum.groupby('Ano')['Volume'].sum()
volume_anual_xrp = dados_xrp.groupby('Ano')['Volume'].sum()
volume_anual_cardano = dados_cardano.groupby('Ano')['Volume'].sum()
volume_anual_solana = dados_solana.groupby('Ano')['Volume'].sum()

# Plotar o crescimento do volume ao longo do tempo
plt.figure(figsize=(12, 6))

plt.plot(volume_anual_bitcoin.index, volume_anual_bitcoin.values, marker='o', label='Bitcoin')
plt.plot(volume_anual_ethereum.index, volume_anual_ethereum.values, marker='o', label='Ethereum')
plt.plot(volume_anual_xrp.index, volume_anual_xrp.values, marker='o', label='XRP')
plt.plot(volume_anual_cardano.index, volume_anual_cardano.values, marker='o', label='Cardano')
plt.plot(volume_anual_solana.index, volume_anual_solana.values, marker='o', label='Solana')

# Configurações do gráfico
plt.title('Crescimento Anual do Volume de Transações de Criptomoedas')
plt.xlabel('Ano')
plt.ylabel('Volume de Transações')
plt.legend()
plt.grid(True)

# Ajustar o layout e exibir
plt.tight_layout()
plt.show()

#Criamos uma coluna 'ano' extraindo o ano da coluna 'data'.
#Usamos groupby('ano')['volume'].sum() para obter o total de volume por ano.
#Plotamos um gráfico de linhas mostrando a evolução do volume ao longo dos anos para cada criptomoeda.
```
